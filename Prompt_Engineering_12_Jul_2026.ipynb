  - With Context
    - You are a medical assistant
3. Input data - User question
4. Output indicator - How should AI present the answer.

```python
from dotenv import load_dotenv, find_dotenv
import os
from openai import OpenAI, OpenAIError
from google.colab import userdata

# Load environment variables
load_dotenv(find_dotenv())

# Read API key from Colab secrets
api_key = userdata.get('OPENAI_API_KEY')

# Create OpenAI client
client = OpenAI(api_key=api_key)

messages = [
    {
        "role": "developer",
        "content": "You are an intelligent assistant."
    }
]

while True:
    user_input = input("Enter some input: ")

    if user_input.lower() == "bye":
        break

    messages.append(
        {
            "role": "user",
            "content": user_input
        }
    )

    try:
        response = client.chat.completions.create(
            model="gpt-4.1-mini",   # or gpt-5-mini, gpt-5
            messages=messages
        )

        reply = response.choices[0].message.content

        print(f"Assistant: {reply}\n")

        messages.append(
            {
                "role": "assistant",
                "content": reply
            }
        )
    except OpenAIError as e:
        print(f"An OpenAI API error occurred: {e}\n")
        # Optionally, you can remove the last user message if the API call failed
        messages.pop()
```
