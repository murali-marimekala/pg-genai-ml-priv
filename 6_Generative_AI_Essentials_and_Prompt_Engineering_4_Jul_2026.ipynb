# Transformers
- It is a deep learning model relies on attention mechanism to processs the data in parallel
- Forms the backbone to latest modles like BERT (encoder) and Decoder (GPT)
- Encoder (BERT - For classification / understanding the input) and decoders (GPT- for text generation)
- Encoder-Decoder - BART
- Eg. You have 100 google reviews of a restaurant.
  - BERT - Encoder -  Does if the reviews are good or bad
  - GPT -  Decoder - It generate a detail resonse
  - BART - Encoder/Decoder - Combines both understanding input and generating text (e.g., summarization, translation)
