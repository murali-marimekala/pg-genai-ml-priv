# Models
1. N-Gram model
2. HMM - Hidden markov models
3. Bayesian Models
4. Word Embedding Models
5. RNN

# When do we choose these models.
1. N-gram model :
  - Predict next word using last few words.
  - Used in Phone texting suggestions.
  - Why ? Very fast , simple , no training on huge data

2. HMM - Hidden Markov models
  - Predict hidden state from observed words
  - Speech recognition
3. Bayesian Models
  - Used probability and prior knowledge
  - Spam email detection
4. Word embedding models
  - Needed when to understand semantic meaning.
  - Convert words into vectors
  - Find similar words - King/Queen - understands semantic meaning of it.
5. RNN (Recurrent Neural Networks)
  - Language translations
6. Transformer models
  - Attention - Looks at the sentence at once using attention
  - Models available in Transformers are
    - Chatgpt
    - Claude
    - Gemini etc
  - Understanding long context, Parallel processing

# SLM vs LLM tradeoff

- Accuracy - LLM
- Inference timing - SLM
- Cost Expensive - LLM
- Deployment - cloud - LLM, SLM - on devices
